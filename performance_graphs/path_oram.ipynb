```python
import matplotlib.pyplot as plt

# Sample data: input sizes and corresponding times

# [2^10, 2^14, 2^20]
input_sizes = [1024, 16384, 131072, 262144, 524288, 1048576]  # y-axis: size of inputs
times = [2.24, 53.97, 496.65, 1061.31, 2222.54 ,4735.25]       # x-axis: time in seconds

plt.figure(figsize=(8,6))
plt.plot(times, input_sizes, marker='o')
plt.xlabel('Time (s)')
plt.ylabel('Input Size')
plt.title('Input Size vs Time (write and then access all data)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.style.use('seaborn-darkgrid')
cool_bg_color = '#0d1b2a'  

#             [2^10   2^14   2^17    2^18    2^19    2^20  ]
input_sizes = [1024, 16384, 131072, 262144, 524288, 1048576]  
times = [2.24, 53.97, 496.65, 1061.31, 2222.54, 4735.25] 

fig, ax = plt.subplots(figsize=(10, 6))
fig.patch.set_facecolor(cool_bg_color)
ax.set_facecolor(cool_bg_color)

ax.plot(times, input_sizes, marker='o', markersize=8, linewidth=2, color='#2a9d8f', label='ORAM Performance')

ax.set_xlabel('Time (s)', fontsize=14, color='white')
ax.set_ylabel('Input Size', fontsize=14, color='white')
ax.set_title('Input Size vs Time (Write and Access All Data)', fontsize=16, color='white')

legend = ax.legend(fontsize=12)
for text in legend.get_texts():
    text.set_color("white")
ax.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray')
plt.tight_layout()
plt.show()
```
